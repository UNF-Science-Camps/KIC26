# P2: Variabler — computerens huskesedler

En **variabel** er en navngivet kasse, som computeren gemmer en værdi i. Man putter noget i
kassen med `=`:

````python
navn = "Pikachu"
hp = 35
print(navn, "har", hp, "HP")
````

Læg mærke til printen: `print` kan få **flere ting adskilt med komma** og sætter selv
mellemrum imellem. Den stil — `print("x er:", x)` — er vores standardmåde at printe på i denne
og næste notebook.

Og så skal `=` læses som *"gem i kassen"* — **ikke** som et matematisk "er lig med". Til
venstre står kassens navn, til højre det, der skal gemmes.

**Navneregler** (Python er striks):

| Navn | OK? | Hvorfor |
|---|---|---|
| `hp` | ✅ | kort og klart |
| `mit_tal` | ✅ | mellemrum erstattes med `_` |
| `mit tal` | ❌ | mellemrum er forbudt i navne |
| `2hurtig` | ❌ | må ikke starte med et tal |
| `hoejde` | ✅ | æ/ø/å undgår vi i navne — skriv `ae/oe/aa` (tekst i citationstegn må stadig alt) |

# P3: Matematik — Python som lommeregner

Python er en glimrende lommeregner — her er hele værktøjskassen:

| Operator | Betydning | Eksempel | Resultat |
|---|---|---|---|
| `+` | plus | `7 + 3` | `10` |
| `-` | minus | `7 - 3` | `4` |
| `*` | gange | `7 * 3` | `21` |
| `/` | division | `7 / 2` | `3.5` |
| `**` | potens | `2 ** 5` | `32` |
| `//` | heltalsdivision (rund ned) | `7 // 2` | `3` |
| `%` | modulo (resten) | `7 % 2` | `1` |

De tre nederste er nye for de fleste: `**` er "opløftet i" ($2^5 = 32$), `//` deler og smider
decimalerne væk, og `%` giver **resten** ved division. Modulo lyder underlig, men er
overraskende nyttig — fx afslører `tal % 2`, om et tal er lige (rest 0) eller ulige (rest 1).

````python
print("7 / 2  giver", 7 / 2)
print("7 // 2 giver", 7 // 2)
print("7 % 2  giver", 7 % 2)
print("2 ** 5 giver", 2 ** 5)
````

## Regnerækkefølge, afrunding og kvadratrødder

Python følger de sædvanlige regneregler: `**` før `*` og `/`, som kommer før `+` og `-` — og
**parenteser vinder over alt**. Er man i tvivl, sætter man en parentes; det koster ingenting.

To små tricks oveni:
- `round(tal, n)` runder af til n decimaler.
- Kvadratrod? Der er ingen √-knap — men det gør ikke noget, for $\sqrt{x} = x^{0.5}$, så `x ** 0.5` klarer opgaven.

````python
print("2 + 3 * 4   giver", 2 + 3 * 4)
print("(2 + 3) * 4 giver", (2 + 3) * 4)
print("Kvadratroden af 2 er", 2 ** 0.5)
print("...eller lidt pænere:", round(2 ** 0.5, 3))
````

# P4: Sandt eller falsk — sammenligninger

Computere træffer beslutninger ved at stille spørgsmål, der kan besvares med **sandt**
(`True`) eller **falsk** (`False`) — bool-værdierne fra P2.6. Spørgsmålene stilles med
**sammenligningsoperatorer**:

| Udtryk | Spørger |
|---|---|
| `a == b` | er a lig med b? |
| `a != b` | er a forskellig fra b? |
| `a < b` | er a mindre end b? |
| `a > b` | er a større end b? |
| `a <= b` | er a mindre end eller lig b? |
| `a >= b` | er a større end eller lig b? |

> ⚠️ **Advarsel — ugens vigtigste tegn-detalje:** ét lighedstegn `=` GEMMER i en variabel
> (P2). To lighedstegn `==` SPØRGER "er de ens?". At bytte rundt på dem er formentlig den
> mest begåede fejl i programmeringshistorien — I får selv lov at rette den i P5. 😈

````python
alder = 17
print("Er du myndig?  ", alder >= 18)
print("Er du 17?      ", alder == 17)
print("Er du IKKE 18? ", alder != 18)
````

## Kombinér spørgsmål: and, or og not

`and` kræver, at BEGGE dele er sande. `or` er tilfreds med mindst én. Og `not` vender
resultatet om: `not True` er `False`.

| a | b | `a and b` | `a or b` |
|---|---|---|---|
| `True` | `True` | `True` | `True` |
| `True` | `False` | `False` | `True` |
| `False` | `False` | `False` | `False` |

````python
temperatur = 23
solskin = True

print("Strandvejr?", temperatur > 20 and solskin)
print("Sofavejr?  ", not (temperatur > 20 and solskin))
````

# P5: if / elif / else — beslutninger

Nu får koden en vilje: `if` kører kun sine linjer, **hvis** en betingelse er sand.

```python

    gør det her
```

To ting er livsvigtige:

- **Kolonet** `:` for enden af betingelsen.
- **Indrykningen**: linjerne, der hører til if'en, rykkes ind (fire mellemrum — eller ét tryk
  på Tab). Indrykningen er IKKE pynt: det er sådan, Python ved, hvilke linjer der hører til
  if'en, og hvor resten af programmet fortsætter. At glemme den er begynderfejl nummer ét —
  I får selv lov at rette den i opgaverne. 😈

````python
alder = 17

if alder >= 18:
    print("Du må stemme til folketingsvalget")
if alder < 18:
    print("Ikke stemme endnu — men machine learning må man heldigvis gerne!")
````

## else og elif

I stedet for selv at stille det modsatte spørgsmål kan man bruge `else` ("ellers"). Og med
`elif` ("ellers, hvis") kan man kæde flere spørgsmål sammen:

````python
point = 73

if point >= 90:
    print("Kæmpe tal! 🤩")
elif point >= 50:
    print("Mellemstort tal")
else:
    print("Lille tal")
````

Python prøver betingelserne **oppefra og ned** og tager den FØRSTE, der er sand — resten
springes over. Som diagram:

```
            point
              │
        point >= 90? ── ja ──►  "Kæmpe tal! 🤩"
              │ nej
        point >= 50? ── ja ──►  "Mellemstort tal"
              │ nej
              ▼
         "Lille tal"
```

Rækkefølgen betyder alt: stod `point >= 50` øverst, ville 95 point aldrig nå frem til "Kæmpe
tal!". Dén fælde ligger og lurer på jer i opgave P5.6...

## if inde i if

Man må gerne putte en `if` **inde i** en anden — så rykker man bare ind én gang til:

````python
vejr = "sol"
temperatur = 25

if vejr == "sol":
    if temperatur > 20:
        print("Strandtur! 🏖️")
    else:
        print("Solbriller på — men behold trøjen")
else:
    print("Inde-dag med kakao og kode ☕")
````

> 🤖 **ML-hjørnet:** der findes en hel familie af ML-modeller — **beslutningstræer** — som
> bogstaveligt talt bare er en stor stak if/else-sætninger. Forskellen på dem og jeres kode
> her: computeren har selv fundet reglerne ud fra data. I opgave P5.8 bygger I den håndlavede
> udgave.
