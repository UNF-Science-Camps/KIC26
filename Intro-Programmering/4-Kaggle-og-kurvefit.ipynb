# P19: curve_fit — lad computeren finde kurven

Kan I huske Opgave P11.1, "den manuelle regressionsmaskine"? I afprøvede en håndfuld
a-værdier, målte fejlen for hver, og valgte den bedste med øjnene. Det virkede — men
det var langsomt, og hvem siger, at det bedste af JERES gæt er den bedst *mulige*
linje?

Mød nu funktionen, der gør præcis det samme — bare automatisk, lynhurtigt og optimalt:
**`curve_fit`**. Den bor i **scipy**, numpys storebror: et bibliotek propfyldt med
færdige videnskabsværktøjer, hvoraf vi i dag kun skal bruge dette ene:

````python
from scipy.optimize import curve_fit
````

## Opskriften

`curve_fit` skal bruge to ting. Først en **modelfunktion**: en helt almindelig
P9-funktion, der beskriver den kurveform, vi *tror* på. Én regel er hellig: **x skal
være funktionens første argument** — og bagefter kommer de parametre, `curve_fit` må
skrue på:

````python
def linje(x, a, b):
    return a * x + b
````

Og så selvfølgelig **dataene**. Mere skal der faktisk ikke til — `curve_fit` finder nu
selv de `a` og `b`, der får linjen til at passe bedst muligt på `dat1`:

````python
par, cov = curve_fit(linje, x1, y1)
print(par)
````

Der kommer to ting retur. `par` er de bedste parametre, i samme rækkefølge som i
funktionen — altså `[a, b]`. Og `cov`? Den beskriver *usikkerheden* på `a` og `b` —
det vender regressions-emnet tilbage til. I dag bruger vi kun `par`.

Lad os pakke parametrene ud og lægge fittet oven på dataene:

````python
a, b = par

xs = np.linspace(0, 10, 100)       # 100 jævnt fordelte x-værdier → en glat linje
plt.scatter(x1, y1, s=10, label="data")
plt.plot(xs, linje(xs, a, b), color="red", label="fit")
plt.title(f"dat1: bedste linje er y = {a:.2f}x + {b:.2f}")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
````

Smukt. 🤌 Og læg mærke til, hvad I lige har gjort: **data + modelfunktion →
computeren finder selv de bedste parametre**. Det er bogstaveligt talt dét, "at træne
en model" betyder — I har officielt fittet jeres første model!

##### Opgave P19.6
Hvor god er en model egentlig? Det måler man med **gennemsnitsfejlen** — jeres egen
funktion fra P9, og præcis det tal, ML-folk kalder **loss**. `curve_fit`s eneste
mission i livet er at gøre loss så lille som muligt, og i Intro-ML kommer I til at
stirre på loss-tal konstant — så lad os hilse ordentligt på. Kør cellen, og ændr så
`a_gaet` og `b_gaet`: hvor tæt kan I komme på `curve_fit`s loss med håndkraft? Kan I
slå den? 😏

````python
def gennemsnitsfejl(gaets, facits):
    return np.mean((gaets - facits)**2)

par, cov = curve_fit(linje, x1, y1)
a_bedst, b_bedst = par

a_gaet, b_gaet = 1, 0      # ← ændr de to tal og jag curve_fits loss!

print(f"curve_fit: {gennemsnitsfejl(linje(x1, a_bedst, b_bedst), y1):10.2f}")
print(f"Jeres gæt: {gennemsnitsfejl(linje(x1, a_gaet, b_gaet), y1):10.2f}")
````
