# P13: Klasser — byggeplaner for objekter

Nu kommer notebookens vigtigste afsnit. 🏗️ I har allerede brugt **objekter** hele ugen uden
at tænke over det: en liste er et objekt, og når I skriver `min_liste.append(7)`, kalder I en
funktion, der *bor på* listen. Sådan en funktion kaldes en **metode**.

I dag bygger I selv den slags. En **klasse** er en *byggeplan*: den beskriver, hvilke data et
objekt skal huske (**attributter**), og hvad det skal kunne (**metoder**). Ud fra én byggeplan
kan man lave lige så mange objekter, man vil.

Og hvorfor skal I kunne det? Fordi **alle neurale netværk i PyTorch er skrevet som klasser**.
Når I om få dage ser `class Net(nn.Module):`, skal I nikke genkendende. 😎

## Den første klasse

````python
class Pokemon:
    def __init__(self, navn, hp):
        self.navn = navn      # en attribut: noget objektet husker
        self.hp = hp          # endnu en attribut

pikachu = Pokemon("Pikachu", 35)
snorlax = Pokemon("Snorlax", 160)

print(pikachu.navn, "har", pikachu.hp, "HP")
print(snorlax.navn, "har", snorlax.hp, "HP")
````

Tre ting at lægge mærke til:

- `class Pokemon:` starter byggeplanen (med kolon og indrykning, ligesom `if` og `def`).
- `__init__` er en særlig metode, der **kører automatisk**, når man laver et nyt objekt med
  `Pokemon("Pikachu", 35)`. Navnet er med to bundstreger i hver ende — Python-slang for
  "det her er specielt".
- `self` betyder **objektet selv**. `self.navn = navn` betyder: "gem argumentet `navn` som en
  attribut på netop DETTE objekt". Derfor kan `pikachu` og `snorlax` huske hver deres navn.

`self` skal stå som første argument i ALLE metoder — men Python udfylder det selv, når man
kalder: `pikachu.navn` og ikke `Pokemon.navn(pikachu)`.

## Metoder: hvad kan objektet?

````python
class Pokemon:
    def __init__(self, navn, hp):
        self.navn = navn
        self.hp = hp

    def praesenter(self):
        print("Jeg hedder", self.navn, "og har", self.hp, "HP!")

    def angrib(self, anden):
        print(self.navn, "angriber", anden.navn, "!")

pikachu = Pokemon("Pikachu", 35)
snorlax = Pokemon("Snorlax", 160)

pikachu.praesenter()
pikachu.angrib(snorlax)
````

## Nedarvning: klasser kan bygge videre på klasser

Man kan lave en ny klasse, der **arver** alt fra en anden og tilføjer sit eget. Skriv
moderklassen i parentes — og kald `super().__init__(...)` for at genbruge dens opstart:

````python
class IldPokemon(Pokemon):              # arver ALT fra Pokemon
    def __init__(self, navn, hp, grader):
        super().__init__(navn, hp)      # koer foerst Pokemons __init__ (gemmer navn og hp)
        self.grader = grader            # ... og tilfoej saa en ny attribut

    def spy_ild(self):
        print(self.navn, "spyr ild på", self.grader, "grader! 🔥")

charmander = IldPokemon("Charmander", 39, 900)
charmander.praesenter()    # metoden er arvet fra Pokemon — gratis!
charmander.spy_ild()       # ... og denne er dens egen
````

Linje for linje:

- `class IldPokemon(Pokemon):` — ny klasse, der kan alt, hvad `Pokemon` kan.
- `super().__init__(navn, hp)` — "kør min moderklasses `__init__` først". Så bliver `navn` og
  `hp` gemt, præcis som for alle andre Pokémon.
- `self.grader = grader` — og herefter tilføjer vi det nye.

## Hvorfor alt det her? Fordi PyTorch! 🤖

Nedenfor står en klasse, der er *formet* præcis som et neuralt netværk i PyTorch — men uden
noget PyTorch i. Læs den — I skal **ikke** kunne skrive den, kun genkende delene:

````python
class Grundmodel:                       # staar i rollen som PyTorchs nn.Module
    def __init__(self):
        print("(Grundmodel goer sig klar...)")

class MinModel(Grundmodel):             # i PyTorch: class MinModel(nn.Module):
    def __init__(self, a, b):
        super().__init__()              # ALTID foerste linje i en PyTorch-model
        self.a = a                      # modellens "viden" gemmes som attributter
        self.b = b

    def forward(self, x):               # forward = "send data igennem modellen"
        return self.a * x + self.b

model = MinModel(3, 4)
print("model.forward(10) giver:", model.forward(10))
````

Genkend delene: en klasse, der arver fra en grundklasse — `super().__init__()` som første
linje — attributter i `__init__` — og en `forward`-metode, der sender data igennem. Det er
skabelonen for **hvert eneste** neurale netværk, I kommer til at se på campen (og de seks
klasser i LLM-workshoppen er bygget over præcis samme læst). Den rigtige `nn.Module` møder I
i Intro-ML notebook 3.

# P14: f-strings — flotte udskrifter

Indtil nu har I printet med kommaer: `print("navn:", navn)`. Det virker fint — men nu lærer I
den måde, resten af campen (og resten af Python-verdenen) printer på: **f-strings**.

Opskriften: sæt et `f` LIGE FØR anførselstegnet, og skriv variabler i krøllede parenteser:

````python
navn = "Pikachu"
hp = 35

print(f"{navn} har {hp} HP")                # f-string: variabler direkte i teksten
print(f"Om 3 år har {navn} {hp + 3} HP?")   # hele UDTRYK maa ogsaa staa i tuborgklammerne
````

Alt mellem `{` og `}` bliver *beregnet* og sat ind i teksten. Det gør f-strings perfekte til
at blande tekst og tal — uden komma-jonglering.

## Formatering: styr på decimalerne

Det virkelig smarte gemmer sig EFTER et kolon i klammerne — en lille formatkode:

| Kode | Eksempel | Gør |
|---|---|---|
| `:.2f` | `f"{pi:.2f}"` → `3.14` | afrund til 2 decimaler |
| `:.3f` | `f"{tab:.3f}"` → `0.457` | 3 decimaler (campens loss-favorit) |
| `:5d` | `f"{n:5d}"` → `   42` | heltal, der fylder 5 pladser (pænt til tabeller) |
| `:>8` | `f"{s:>8}"` → `   kat` | højrestil tekst på 8 pladser |
| `:.1%` | `f"{andel:.1%}"` → `97.5%` | vis et 0-1-tal som procent |

````python
pi = 3.14159265
skridt = 42
tab = 0.456789

print(f"pi med to decimaler: {pi:.2f}")
print(f"skridt som pæn kolonne: {skridt:5d}")
print(f"loss med tre decimaler: {tab:.3f}")
````

# P15: Ordbøger — opslagsværker (og en tokenizer!)

Sidste nye datatype — og den har en hemmelig dagsorden. 🕵️

Her er problemet, den skal løse: en sprogmodel (som den, I skal træne i LLM-workshoppen) kan
**kun regne på tal**. Den kan ikke se bogstavet `"k"` — den skal bruge et *tal*, der betyder
`"k"`. Vi mangler altså et **opslagsværk** fra tegn til tal. Det hedder i Python en **ordbog**
(engelsk: *dictionary* eller bare `dict`):

````python
dansk_til_engelsk = {"kat": "cat", "hund": "dog", "abe": "monkey"}

print(dansk_til_engelsk["kat"])       # slaa op med kantede parenteser
print(dansk_til_engelsk["abe"])
````

En ordbog består af **nøgler** (det man slår op PÅ) og **værdier** (det man får tilbage) —
skrevet som `{noegle: vaerdi, noegle: vaerdi}`. Man kan tilføje nye opslag, tjekke om en
nøgle findes med `in`, og loope over nøglerne:

````python
dansk_til_engelsk["fisk"] = "fish"                # tilfoej et nyt opslag

print("fisk" in dansk_til_engelsk)                # findes noeglen?
print("giraf" in dansk_til_engelsk)

for ord in dansk_til_engelsk:                     # en for-loekke besoeger noeglerne
    print(ord, "hedder", dansk_til_engelsk[ord], "på engelsk")
````

## Tokenizeren: tegn bliver til tal

Nu til den hemmelige dagsorden. Vi bygger to ordbøger: `stoi` ("string to integer") oversætter
tegn til tal, og `itos` ("integer to string") oversætter tilbage. Præcis de to navne bruger
LLM-workshoppen også:

````python
stoi = {"a": 0, "b": 1, "e": 2, "k": 3, "t": 4}   # tegn -> tal
itos = {0: "a", 1: "b", 2: "e", 3: "k", 4: "t"}   # tal -> tegn (samme, spejlvendt)

print("k bliver til tallet:", stoi["k"])
print("tallet 2 bliver til:", itos[2])
````

Og skal en hel liste af tegn limes sammen til én tekst igen, bruger man `"".join(...)`:

````python
tegn_liste = ["k", "a", "t"]
print("".join(tegn_liste))            # limer listens tegn sammen til en streng
````

## Én linje, I skal kunne GENKENDE (ikke skrive)

I LLM-workshoppen bygges `stoi` automatisk ud fra alfabetet med denne linje:

> 🤖 **ML-hjørnet:** tegn → tal → *regn løs* → tal → tegn. Det er HELE fundamentet under
