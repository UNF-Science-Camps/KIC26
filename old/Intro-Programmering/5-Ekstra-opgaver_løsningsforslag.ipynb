### Opgave E.1 — Skudårs-fabrikken 📅

Et år er skudår, hvis det er deleligt med 4 — **undtagen** hele hundredår, som kun er
skudår, hvis de også er delelige med 400. År 2000 var altså skudår, men år 1900 var ikke!

Funktionen `er_skudaar` er allerede bygget (genkender I det boolske udtryk fra notebook 1?).
**Jeres mission:** brug den i en løkke over alle år fra 1900 til 2025, saml skudårene i en
liste, og gør f-string-rapporten færdig.

Forventet output:
```
Der er 31 skudår mellem 1900 og 2025
Det første er 1904 og det sidste er 2024
```

⭐ Ekstra: hvor mange skudår er der mellem år 1 og år 3000?

````python
def er_skudaar(aar):
    return (aar % 4 == 0 and aar % 100 != 0) or (aar % 400 == 0)

antal = 0
skudaar_liste = []
for aar in range(1900, 2026):        # husk: 2026 er IKKE med
    if er_skudaar(aar):
        antal = antal + 1
        skudaar_liste.append(aar)

print(f"Der er {antal} skudår mellem 1900 og 2025")
print(f"Det første er {skudaar_liste[0]} og det sidste er {skudaar_liste[-1]}")

# ⭐ Ekstra: hvor mange skudår er der mellem år 1 og år 3000?
antal_alle = 0
for aar in range(1, 3001):
    if er_skudaar(aar):
        antal_alle = antal_alle + 1
print(f"Mellem år 1 og år 3000 er der {antal_alle} skudår")
````

<span style="color:red"><b>LØSNINGSFORSLAG:</b> Hullet i løkken er <code>if er_skudaar(aar):</code>, og rapporten indekserer med <code>skudaar_liste[0]</code> og <code>skudaar_liste[-1]</code> (målt: 31 skudår, første 1904, sidste 2024 — bemærk at 1900 netop IKKE er med, fordi det er et hundredår). ⭐-delen genbruger løkken med <code>range(1, 3001)</code> (målt: 727 skudår).</span>

### Opgave E.2 — Primtals-detektoren 🔎

Et **primtal** er et helt tal over 1, der kun er deleligt med 1 og sig selv (2, 3, 5, 7, 11...).
Strategien: prøv ALLE mulige divisorer, og brug en **flag-variabel** til at huske, om vi
fandt én, der gik op. Flaget starter på `False` og bliver kun vippet til `True`, hvis vi
fanger noget — og så kører løkken bare færdig, det gør ikke noget.

**Jeres mission:** udfyld flag-linjen i funktionen, og brug den bagefter til at finde alle
primtal op til 100.

Forventet output:
```
True False
Der er 25 primtal under 100
[2, 3, 5, 7, 11, ...]
```

````python
def er_primtal(n):
    if n < 2:
        return False                 # 0 og 1 er ikke primtal
    fandt_divisor = False            # flag-variablen: har vi fundet en divisor endnu?
    for divisor in range(2, n):
        if n % divisor == 0:
            fandt_divisor = True     # vip flaget!
    return not fandt_divisor

print(er_primtal(7), er_primtal(10))     # forventet: True False

primtal = []
for n in range(2, 101):
    if er_primtal(n):
        primtal.append(n)

print(f"Der er {len(primtal)} primtal under 100")
print(primtal)
````

<span style="color:red"><b>LØSNINGSFORSLAG:</b> Flag-linjen er <code>fandt_divisor = True</code>, og løkken nederst er <code>if er_primtal(n): primtal.append(n)</code> (målt: 25 primtal under 100). (Lærernote: pointen med flag-variablen er, at løkken bare kører færdig uden tidlig exit — eleverne har med vilje ikke lært at afbryde en løkke, og det er heller ikke nødvendigt. Hurtige elever kan spørge, om man kan nøjes med divisorer op til kvadratroden — ja, men det er ren bonus.)</span>

### Opgave E.3 — Gæt-tallet turbo 🎯

Computeren skal gætte det hemmelige tal — men i stedet for at rykke 1 ad gangen (som i
notebook 1 og 2) bruger den turbo-strategien: **start med et kæmpe skridt, og halvér det
efter hver runde**. I ML hedder skridtstørrelsen *learning rate* — og tricket med at
starte stort og slutte småt er præcis sådan, man træner rigtige modeller.

**Jeres mission:** udfyld de to huller i strategien, kør, og nyd loggen + plottet.
Bemærk at fejlen godt kan hoppe OP i en enkelt runde (kig på runde 2!) — skridtene bliver
mindre, så den fanger sig selv igen.

Forventet output (starten):
```
Runde  0: gæt =    0.00, fejl =  73.00, skridt = 64.00
Runde  1: gæt =   64.00, fejl =   9.00, skridt = 32.00
Runde  2: gæt =   96.00, fejl =  23.00, skridt = 16.00
```

⭐ Ekstra: prøv et andet hemmeligt tal — og prøv at gange skridtet med 0.9 i stedet for at
halvere. Hvad sker der med kurven?

````python
hemmeligt_tal = 73.0
gaet = 0.0
skridt = 64.0                        # startskridtet — det skal halveres efter hver runde!

fejl_liste = []
for runde in range(12):
    fejl = abs(gaet - hemmeligt_tal)             # abs = tallet uden fortegn
    fejl_liste.append(fejl)
    print(f"Runde {runde:2d}: gæt = {gaet:7.2f}, fejl = {fejl:6.2f}, skridt = {skridt:.2f}")
    if gaet < hemmeligt_tal:
        gaet = gaet + skridt         # for lavt → gå op
    else:
        gaet = gaet - skridt         # for højt (eller spot on) → gå ned
    skridt = skridt / 2              # halvér skridtet

plt.plot(fejl_liste, marker="o")
plt.xlabel("runde")
plt.ylabel("|fejl| (hvor langt fra målet?)")
plt.title("Fejlen styrtdykker — ligesom en loss-kurve!")
plt.show()
````

<span style="color:red"><b>LØSNINGSFORSLAG:</b> Hullerne er <code>gaet = gaet - skridt</code> og <code>skridt = skridt / 2</code> (målt: efter 12 runder er fejlen nede på 0.06 — og i runde 2 hopper den op på 23, fordi det store skridt skyder forbi målet). (Lærernote: dette ER learning rate-pointen: for store skridt → man skyder forbi, for små → sneglefart. Med faktor 0.9 i stedet for halvering falder fejlen meget langsommere, men mere "blødt". Analogien vender tilbage som gradient descent i Intro-ML notebook 2.)</span>

### Opgave E.4 — Vejrstationen 🌡️

I har fået et helt års temperaturmålinger (simuleret — men med seed 42, så alle får
præcis de samme). **Jeres mission:** udfyld statistikkerne med numpy, gør rapporten
færdig, og nyd histogrammet.

Forventet output (starten):
```
Gennemsnit:  18.0 °C
Varmeste dag: 37.3 °C
```

⭐ Ekstra-missionen står i bunden af cellen — og prøv bagefter at ændre gennemsnittet fra

### Opgave E.5 — Tekst-statistik 📖

På campens sidste dag skal I træne en sprogmodel på eventyr-tekst — og det allerførste
skridt i AL sprog-ML er at **kende sine tegn og tælle dem** (tokenizeren fra notebook 3
siger hej 👋). Her er et lille eventyr at øve sig på.

**Jeres mission:** tæl bogstaverne med en ordbog, og find derefter top-5 — HELT uden
smarte imports: fem gange i træk finder I det hyppigste bogstav, der ikke allerede har
fået en plads på skamlen.

Forventet output (starten):
```
Teksten har 541 tegn og 26 forskellige bogstaver
1. plads: 'e' med 77 forekomster
```

⭐ Ekstra: tegn top-5 som søjlediagram (hullerne står klar nederst).

````python
tekst = """Ridderen Asbjørn red gennem tågeskoven med sit rustne sværd og et halvt rugbrød i sadeltasken.
Ved den gamle stenbro ventede druiden Yrsa, som kunne tale med både ravne og regnorme.
Under broen boede trolden Grombuk, der ikke havde sovet i tre hundrede år og derfor var i exceptionelt dårligt humør.
Da månen stod højest, viste druiden ridderen en hemmelig rune, der lyste blåt i mørket.
Trolden brølede, så broen rystede, men Asbjørn løftede sværdet, og runen sang.
Og fra den nat fik skoven fred, mens de tre delte rugbrødet i månens skær."""

# 1) Tæl hvert bogstav med en ordbog (.lower() slår store og små bogstaver sammen)
taeller = {}
for tegn in tekst.lower():
    if tegn in "abcdefghijklmnopqrstuvwxyzæøå":      # kun bogstaver — ingen mellemrum og punktummer
        if tegn in taeller:
            taeller[tegn] = taeller[tegn] + 1    # tegnet er set før: tæl én op
        else:
            taeller[tegn] = 1                        # første gang vi ser tegnet

print(f"Teksten har {len(tekst)} tegn og {len(taeller)} forskellige bogstaver")

# 2) Top-5: find fem gange i træk det hyppigste bogstav, der ikke allerede er på skamlen
top_bogstaver = []
top_antal = []
for plads in range(5):
    bedste_tegn = ""
    bedste_antal = 0
    for tegn in taeller:
        if taeller[tegn] > bedste_antal and tegn not in top_bogstaver:
            bedste_tegn = tegn
            bedste_antal = taeller[tegn]         # opdatér rekorden
    top_bogstaver.append(bedste_tegn)
    top_antal.append(bedste_antal)
    print(f"{plads + 1}. plads: '{bedste_tegn}' med {bedste_antal} forekomster")

# 3) ⭐ Søjlediagram over top-5:
plt.bar(top_bogstaver, top_antal)
plt.ylabel("forekomster")
plt.title("De 5 mest brugte bogstaver")
plt.show()
````

<span style="color:red"><b>LØSNINGSFORSLAG:</b> Hullerne er <code>taeller[tegn] = taeller[tegn] + 1</code>, <code>bedste_antal = taeller[tegn]</code> og <code>plt.bar(top_bogstaver, top_antal)</code> (målt top-5: 'e' 77, 'r' 46, 'd' 40, 'n' 39, 't' 29 — ud af 541 tegn og 26 forskellige bogstaver). (Lærernote: 'e' vinder i stort set al dansk tekst — nævn gerne, at det er præcis den slags tælle-statistik, en sprogmodel bygger på, og at eleverne møder eventyr-tekstdata igen i LLM-workshoppen.)</span>

### Opgave E.6 — Klasse-RPG ⚔️

Pikachu mod Snorlax, ti runder, ingen nåde. Klassen `Pokemon` har en `angrib`-metode, der
lægger et terningkast oven i styrken — og hele kampen logges med f-strings. En
**flag-variabel** sørger for, at kampen stopper, når nogen rammer 0 HP (løkken kører bare
tomme runder færdig — det gør ikke noget).

**Jeres mission:** skab Snorlax (80 HP, styrke 5), og udfyld sejrs-beskeden for det
tilfælde, at Pikachu taber.

Forventet output (starten):
```
--- Runde 1 ---
Pikachu angriber! Snorlax mister 12 HP og har 68 tilbage
Snorlax angriber! Pikachu mister 9 HP og har 51 tilbage
```

⭐ Ekstra: justér styrkerne, så kampen bliver så tæt som muligt — og giv klassen en ekstra
attribut `angreb_navn` ("Tordenchok!", "Mavesjask!"), som bruges i loggen.

````python
class Pokemon:
    def __init__(self, navn, hp, styrke):
        self.navn = navn
        self.hp = hp
        self.styrke = styrke

    def angrib(self, modstander):
        skade = self.styrke + np.random.randint(0, 6)          # styrke + terningkast (0-5)
        modstander.hp = modstander.hp - skade
        print(f"{self.navn} angriber! {modstander.navn} mister {skade} HP og har {modstander.hp} tilbage")

np.random.seed(42)                       # samme kamp for alle
pikachu = Pokemon("Pikachu", 60, 9)
snorlax = Pokemon("Snorlax", 80, 5)      # skab Snorlax: 80 HP, styrke 5

kampen_er_slut = False                   # flag-variablen
for runde in range(1, 11):
    if not kampen_er_slut:
        print(f"--- Runde {runde} ---")
        pikachu.angrib(snorlax)
        if snorlax.hp <= 0:
            print(f"{snorlax.navn} er besejret! {pikachu.navn} vinder! 🏆")
            kampen_er_slut = True
    if not kampen_er_slut:
        snorlax.angrib(pikachu)
        if pikachu.hp <= 0:
            print(f"{pikachu.navn} er besejret! {snorlax.navn} vinder! 🏆")
            kampen_er_slut = True

if not kampen_er_slut:
    print("Tiden er gået — dommerne erklærer kampen uafgjort! 🤝")
````

<span style="color:red"><b>LØSNINGSFORSLAG:</b> <code>snorlax = Pokemon("Snorlax", 80, 5)</code>, og sejrs-beskeden spejler Pikachus: print + <code>kampen_er_slut = True</code> (målt: Pikachu vinder i runde 7 — med kun 13 HP tilbage, så det er tættere, end det ser ud!). ⭐-delen: den nye attribut sættes i <code>__init__</code> ligesom de andre (<code>self.angreb_navn = angreb_navn</code>) og flettes ind i angrib-loggens f-string.</span>

### Opgave E.7 — Fejl-detektiven jr. 🕵️ (boss-level 🐛×5)

Karakter-maskinen nedenfor skal lave statistik over klassens quiz-point — men der gemmer
sig **FEM fejl** i den. Nogle crasher med en fejlbesked, én snyder i **total stilhed**.
Find og ret alle fem!

*Tip: der er én fejl i hver af de tre funktioner og to i rapporten nederst. Ret én ad
gangen, kør igen, og lad fejlbeskederne guide jer (læs dem nedefra!). Den stille fejl
fanger I kun ved at regne efter i hovedet.*

Når alt er rettet, skal outputtet være:
```
Gennemsnit: 63.2 point
Bedste resultat: 92 point
Sidste elev fik: 74 point
Antal beståede: 7
```

````python
# Karakter-maskinen 3000 — nu med alle FEM fejl rettet ✅
point = [92, 71, 45, 88, 12, 67, 90, 55, 38, 74]

def gennemsnit(tal_liste):                                 # FEJL 1: kolon manglede
    total = 0
    for tal in tal_liste:
        total = total + tal
    return total / len(tal_liste)

def find_bedste(tal_liste):
    bedste = 0
    for i in range(0, len(tal_liste)):                     # FEJL 2: startede ved 1 og sprang plads 0 over — dér lå 92!
        if tal_liste[i] > bedste:
            bedste = tal_liste[i]
    return bedste

def tael_bestaaede(tal_liste):
    antal = 0
    for tal in tal_liste:
        if tal >= 50:
            antal = antal + 1                              # FEJL 3: indrykningen manglede
    return antal

print(f"Gennemsnit: {gennemsnit(point):.1f} point")
print(f"Bedste resultat: {find_bedste(point)} point")
print(f"Sidste elev fik: {point[len(point) - 1]} point")   # FEJL 4: plads len(point) findes ikke — sidste plads er len - 1
print(f"Antal beståede: {tael_bestaaede(point)}")          # FEJL 5: tekst + tal crasher — f-strings klarer det
````

<span style="color:red"><b>LØSNINGSFORSLAG:</b> De fem fejl: (1) manglende kolon efter <code>def gennemsnit(tal_liste)</code> — SyntaxError; (2) <code>range(1, len(tal_liste))</code> springer plads 0 over — DEN stille fejl: bedste resultat bliver 90 i stedet for 92; (3) manglende indrykning under <code>if tal >= 50:</code> — IndentationError; (4) <code>point[len(point)]</code> peger én plads forbi listen — sidste plads er <code>len(point) - 1</code> (eller bare <code>point[-1]</code>); (5) <code>"tekst" + tal</code> crasher med TypeError — brug en f-string. Rettet output: 63.2 / 92 / 74 / 7. (Lærernote: eleverne har mødt hver eneste af de fem fejltyper i notebook 1-3 — mind dem om at læse fejlbeskeder nedefra.)</span>

### Opgave E.8 ⭐ — k-NN i ren numpy 🤖

Tid til dagens hovedret: **k-nærmeste-naboer** (k-NN) er en RIGTIG ML-algoritme — og I kan
bygge den med KUN dette emnes værktøjer: numpy, løkker og if-sætninger. Idéen: en ny,
ukendt blomst 🌼 skal klassificeres som rød eller blå. Vi finder de **k nærmeste** kendte
blomster og lader dem **stemme** — flertallet bestemmer!

Trin for trin:
1. **Afstande:** Pythagoras på alle 40 punkter i ét hug (den linje er givet — nyd den).
2. **`np.argsort`** giver punkternes numre sorteret fra tættest til fjernest.
3. **Afstemningen:** en for-løkke tæller stemmerne blandt de k nærmeste.

**Jeres mission:** udfyld de tre huller. Prøv bagefter at flytte den nye blomst og at
skrue på `k` (1? 15?) — hvornår skifter afstemningen?

````python
np.random.seed(42)                                    # samme blomster for alle

# 20 røde og 20 blå blomster, hver art samlet om sit eget centrum:
roede = np.random.normal([2, 2], 0.8, size=(20, 2))   # 20 rækker med (x, y)
blaa = np.random.normal([5, 5], 0.8, size=(20, 2))
punkter = np.concatenate([roede, blaa])               # alle 40 punkter i ét array
labels = [0] * 20 + [1] * 20                          # 0 = rød, 1 = blå (samme rækkefølge!)

ny = np.array([3.4, 3.1])                             # den ukendte blomst
k = 5                                                 # antal naboer, der må stemme

# 1) Afstanden fra ny til ALLE 40 punkter på én gang (Pythagoras, vektoriseret!):
afstande = np.sqrt(np.sum((punkter - ny) ** 2, axis=1))

# 2) np.argsort giver rækkefølgen fra tættest til fjernest:
raekkefoelge = np.argsort(afstande)
naermeste = raekkefoelge[:k]                          # de k første i rækkefølgen (slicing!)

# 3) Flertalsafstemning blandt de k nærmeste:
stemmer_roed = 0
stemmer_blaa = 0
for nummer in naermeste:
    if labels[nummer] == 0:
        stemmer_roed = stemmer_roed + 1               # én stemme til rød
    else:
        stemmer_blaa = stemmer_blaa + 1               # én stemme til blå

if stemmer_roed > stemmer_blaa:
    print(f"Afstemning: {stemmer_roed}-{stemmer_blaa} — den nye blomst er RØD! 🔴")
else:
    print(f"Afstemning: {stemmer_blaa}-{stemmer_roed} — den nye blomst er BLÅ! 🔵")

# 4) Det hele tegnet op:
plt.scatter(roede[:, 0], roede[:, 1], color="crimson", label="røde")
plt.scatter(blaa[:, 0], blaa[:, 1], color="royalblue", label="blå")
plt.scatter(ny[0], ny[1], color="gold", s=250, marker="*", edgecolors="black", label="den nye blomst")
plt.legend()
plt.title(f"k-NN med k = {k}")
plt.show()
````

<span style="color:red"><b>LØSNINGSFORSLAG:</b> Hullerne er <code>naermeste = raekkefoelge[:k]</code> og de to stemme-linjer <code>stemmer_roed = stemmer_roed + 1</code> / <code>stemmer_blaa = stemmer_blaa + 1</code>. (Lærernote: forventet stemmeresultat med k = 5 er (målt: 3-2 til RØD) — punktet ligger bevidst tæt på grænsen; k = 1 giver rød, og k = 15 giver (målt: 9-6) stadig rød. Pointér gerne, at k-NN slet ikke "træner" — al regnekraften ligger i afstandene. Godt diskussionsspørgsmål: hvad sker der med et meget stort k? Svar: flertallet i HELE datasættet vinder altid, uanset hvor punktet ligger.)</span>

### Opgave E.9 ⭐ — Grid-search-regression 🗺️

Den sidste boss: find den bedste rette linje $y = a \cdot x + b$ gennem 1000 datapunkter —
ved at prøve **alle kombinationer** af a og b i et stort gitter og beholde den med den
laveste gennemsnitsfejl. Det hedder *grid search*, og det er "curve_fit på den hårde måde":
i notebook 4 fandt scipy svaret på et splitsekund — nu ser I, hvad der egentlig menes med
"find de bedste parametre".

**Jeres mission:** udfyld fejl-målingen og gem-den-bedste-logikken. (Setup-cellen har
allerede hentet `dat1.txt` til jer.)

Forventet output:
```
Bedste kombination: a = 3.00, b = 4.00 (fejl = 3.956)
```

⭐ Ekstra-tanke: gitteret prøver 60 × 80 = 4800 kombinationer. Et neuralt netværk har
MILLIONER af parametre — hvor mange kombinationer skulle et gitter så prøve? Dét er
derfor, smartere metoder (gradient descent!) findes. I møder dem i Intro-ML.

````python
data = np.loadtxt("dat1.txt")            # 1000 rækker med (x, y)
x = data[:, 0]
y = data[:, 1]

def gennemsnitsfejl(a, b):
    gaets = a * x + b                    # linjens gæt for ALLE x på én gang
    return np.mean((gaets - y) ** 2)     # gennemsnitlig kvadreret fejl — vores loss!

bedste_fejl = 999999.0                   # start med en absurd dårlig rekord
bedste_a = 0.0
bedste_b = 0.0
for a in np.arange(0, 6, 0.1):           # 60 værdier af a...
    for b in np.arange(0, 8, 0.1):       # ...gange 80 værdier af b
        fejl = gennemsnitsfejl(a, b)     # mål fejlen for netop denne kombination
        if fejl < bedste_fejl:
            bedste_fejl = fejl           # ny rekord: gem fejl, a og b
            bedste_a = a
            bedste_b = b

print(f"Bedste kombination: a = {bedste_a:.2f}, b = {bedste_b:.2f} (fejl = {bedste_fejl:.3f})")

x_linje = np.linspace(np.min(x), np.max(x), 100)
plt.scatter(x, y, s=8, alpha=0.4, label="data")
plt.plot(x_linje, bedste_a * x_linje + bedste_b, color="crimson", linewidth=3,
         label=f"bedste linje: y = {bedste_a:.2f}x + {bedste_b:.2f}")
plt.legend()
plt.title("Grid search: 4800 linjer prøvet, én vandt")
plt.show()
````

<span style="color:red"><b>LØSNINGSFORSLAG:</b> Hullerne er <code>fejl = gennemsnitsfejl(a, b)</code> og de tre gem-linjer <code>bedste_fejl = fejl</code>, <code>bedste_a = a</code>, <code>bedste_b = b</code> (målt: a = 3.00, b = 4.00, fejl = 3.956 — curve_fit i notebook 4 fandt a ≈ 2.99 og b ≈ 4.05, så gitteret rammer så godt, som dets 0.1-opløsning tillader). (Lærernote: fin anledning til at pege fremad: gradient descent finder det samme UDEN at prøve alle kombinationer — det er hele pointen i Intro-ML notebook 2.)</span>
