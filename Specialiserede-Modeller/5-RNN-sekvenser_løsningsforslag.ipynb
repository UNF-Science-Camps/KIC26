# 8: Sekvenser — og modstanderen, der skal slås

## Første regel om tidsserier: splittet er KRONOLOGISK

I al tidligere ML blandede vi data og splittede tilfældigt. **Det er FORBUDT her.**
Modellen skal forudsige FREMTIDEN ud fra FORTIDEN — så testsættet skal ligge EFTER
træningssættet i tid. Blander man, træner modellen på tirsdag og onsdag og "forudsiger"
mandagen imellem dem — spådom med facitliste. Vi bruger de første 80 % af dagene til
træning og de sidste 20 % til test.

## Modstanderen: den naive prognose

Før man bygger NOGEN prognosemodel, skal man kende den naive baseline:
**"i morgen = i dag"**. Den kræver nul træning, nul parametre — og den er pinligt
svær at slå på temperaturer:

```python
n_traen = int(0.8 * len(temperatur))

i_gaar = temperatur[n_traen:-1]        # dagens temperatur...
i_morgen = temperatur[n_traen + 1:]    # ...og dagen efter

naiv_mae = np.abs(i_gaar - i_morgen).mean()
print(f"naiv prognose ('i morgen = i dag'): {naiv_mae:.2f} graders fejl i snit")
```

**~1,5 grader i gennemsnitsfejl — uden at løfte en finger.** Det tal skal enhver
model måles imod. En prognosemodel, der er dårligere end "i morgen = i dag", er ikke en
prognosemodel; den er spild af strøm. (Journalist-huskeregel: når nogen praler af en
"87 % præcis AI-prognose" — spørg altid, hvad den naive baseline var!)

## Glidende vinduer: fra serie til træningsdata

Hvordan træner man overhovedet en model på én lang serie? Man klipper den op i
**vinduer**: de sidste 7 dage er input, dag nummer 8 er facit. Så glider man én dag
frem og klipper igen:

```python
vindue = 7
X_liste, y_liste = [], []
for i in range(len(temperatur) - vindue):
    X_liste.append(temperatur[i:i + vindue])       # 7 dage ind...
    y_liste.append(temperatur[i + vindue])         # ...dag 8 er facit

X_alle = np.array(X_liste)
y_alle = np.array(y_liste)
print("vinduer:", X_alle.shape, "| facit:", y_alle.shape)
print("eksempel — input:", X_alle[0].round(1), "→ facit:", round(float(y_alle[0]), 1))
```

```python
n_traen = int(0.8 * len(X_alle))                   # kronologisk split!
X_traen_np, X_test_np = X_alle[:n_traen], X_alle[n_traen:]
y_traen_np, y_test_np = y_alle[:n_traen], y_alle[n_traen:]

# standardisér — med TRÆNINGSSÆTTETS gennemsnit/spredning (testen er "fremtiden"!)
gns, spredning = X_traen_np.mean(), X_traen_np.std()

X_traen = torch.tensor((X_traen_np - gns) / spredning)
X_test = torch.tensor((X_test_np - gns) / spredning)
y_traen = torch.tensor((y_traen_np - gns) / spredning)
y_test_celsius = torch.tensor(y_test_np)           # facit beholdes i rigtige grader

print("træning:", X_traen.shape, "| test:", X_test.shape)
```

## Første udfordrer: et dense-netværk på vinduerne

7 tal ind, 1 tal ud — det er jo bare et almindeligt netværk fra Intro-ML! Bemærk
hvordan vi regner fejlen om til GRADER til sidst (ganger spredningen på igen), så
tallet kan sammenlignes med den naive baseline:

```python
torch.manual_seed(42)
dense_model = nn.Sequential(nn.Linear(vindue, 32), nn.ReLU(), nn.Linear(32, 1))
tabsfunktion = nn.MSELoss()
optimizer = torch.optim.Adam(dense_model.parameters(), lr=0.001)

for epoke in range(1000):
    optimizer.zero_grad()
    tab = tabsfunktion(dense_model(X_traen).squeeze(), y_traen)
    tab.backward()
    optimizer.step()

with torch.no_grad():
    forudsagt = dense_model(X_test).squeeze() * spredning + gns    # tilbage til °C
dense_mae = (forudsagt - y_test_celsius).abs().mean()
print(f"dense-net: {dense_mae.item():.2f} graders fejl  (naiv: {naiv_mae:.2f})")
```

##### Opgave 8.5
Hvor mange dages hukommelse hjælper dense-nettet? Prøv `vindue` på **1, 7 og 30**
(byg vinduerne og modellen forfra hver gang — skabelonen klarer det). Hvor meget
hjælper mere fortid egentlig?

```python
def dense_forsoeg(vindue_str):
    Xl, yl = [], []
    for i in range(len(temperatur) - vindue_str):
        Xl.append(temperatur[i:i + vindue_str])
        yl.append(temperatur[i + vindue_str])
    Xa, ya = np.array(Xl), np.array(yl)
    n = int(0.8 * len(Xa))
    g, sp = Xa[:n].mean(), Xa[:n].std()
    Xt = torch.tensor((Xa[:n] - g) / sp)
    Xv = torch.tensor((Xa[n:] - g) / sp)
    yt = torch.tensor((ya[:n] - g) / sp)
    torch.manual_seed(42)
    model_f = nn.Sequential(nn.Linear(vindue_str, 32), nn.ReLU(), nn.Linear(32, 1))
    opt = torch.optim.Adam(model_f.parameters(), lr=0.001)
    for epoke in range(1000):
        opt.zero_grad()
        tab = tabsfunktion(model_f(Xt).squeeze(), yt)
        tab.backward()
        opt.step()
    with torch.no_grad():
        forudsagt_f = model_f(Xv).squeeze() * sp + g
    return (forudsagt_f - torch.tensor(ya[n:])).abs().mean().item()

for v in [1, 7, 30]:
    print(f"vindue {v:2d}: {dense_forsoeg(v):.2f} grader")
```

<span style="color:red"><b>LØSNINGSFORSLAG:</b> Typisk: vindue 1 ≈ 1,6-1,7 (modellen genopfinder stort set den naive prognose), vindue 7 ≈ 1,5, vindue 30 ≈ 1,5-2,1. Mere fortid hjælper altså kun marginalt — morgendagens temperatur afhænger næsten kun af i dag. Læren: mere input er ikke automatisk bedre; det afhænger af, om informationen ER der.</span>

# 9: LSTM — netværket der læser én dag ad gangen

Dense-nettet ser vinduet som 7 løsrevne tal — det ved ikke, at dag 3 kommer FØR dag 4
(byt to inputs om, og det er ligeglat: det lærer bare nye vægte). En **RNN**
(*recurrent neural network*) gør noget fundamentalt andet: den læser sekvensen ét
skridt ad gangen og bærer en **hukommelse** med sig:

> læs dag 1 → opdatér hukommelsen → læs dag 2 → opdatér hukommelsen →... → gæt

**LSTM** (*Long Short-Term Memory*) er den forbedrede udgave, alle bruger: dens
hukommelsescelle har små indbyggede "porte", der lærer hvad der skal **huskes**, hvad
der skal **glemmes**, og hvad der skal **siges videre**. (Portene er selv små neurale
lag — matematikken bag springer vi over med god samvittighed.)

I PyTorch er det ét lag: `nn.LSTM`. Det spiser 3D-input `(antal, tidsskridt, features)`
— vores vinduer skal altså have en ekstra dimension på (7 dage × 1 tal pr. dag), og vi
skriver **`batch_first=True`** (ellers forventer PyTorch en anden rækkefølge — en
klassisk fælde):

```python
class VejrLSTM(nn.Module):
    def __init__(self, hukommelse=32):
        super().__init__()
        self.lstm = nn.LSTM(input_size=1, hidden_size=hukommelse, batch_first=True)
        self.fc = nn.Linear(hukommelse, 1)

    def forward(self, x):                 # x: (antal, 7 dage, 1 tal)
        hukommelser, _ = self.lstm(x)     # hukommelsen efter HVERT tidsskridt
        sidste = hukommelser[:, -1, :]    # vi bruger den efter SIDSTE dag
        return self.fc(sidste)

X_traen_3d = X_traen.unsqueeze(2)         # (3209, 7) → (3209, 7, 1)
X_test_3d = X_test.unsqueeze(2)
print("LSTM-input:", X_traen_3d.shape, "  ← (antal, tidsskridt, features)")
```

```python
torch.manual_seed(42)
lstm_model = VejrLSTM()
optimizer = torch.optim.Adam(lstm_model.parameters(), lr=0.001)

for epoke in range(60):
    for i in range(0, len(X_traen_3d), 256):
        optimizer.zero_grad()
        tab = tabsfunktion(lstm_model(X_traen_3d[i:i + 256]).squeeze(), y_traen[i:i + 256])
        tab.backward()
        optimizer.step()

with torch.no_grad():
    forudsagt = lstm_model(X_test_3d).squeeze() * spredning + gns
lstm_mae = (forudsagt - y_test_celsius).abs().mean()

print(f"naiv:  {naiv_mae:.2f} grader")
print(f"dense: {dense_mae.item():.2f} grader")
print(f"LSTM:  {lstm_mae.item():.2f} grader")
```

**Dagens ærlige facit: det er (stort set) dødt løb.** Alle tre lander omkring 1,5
grader. Er LSTM'en så til grin? Nej — men temperatur-prognose ER bare et problem, hvor
"i går + støj" næsten er hele sandheden, så der er intet dybt sekvens-mønster at hente.

LSTM'ens superkræfter viser sig, når rækkefølgen bærer RIGTIG information: melodier,
hjerterytmer, sensor-mønstre — og frem for alt **sprog**, hvor ordenes rækkefølge er
alt. Husk denne notebook, når I på campens sidste dag møder sprogmodeller: en sætning
er en sekvens, og "gæt næste ord" er PRÆCIS "gæt morgendagens temperatur" — bare med
ord.

## Prognoser flere dage frem: fejlen der vokser

Én dag frem er én ting — men hvad med 14 dage? Tricket: brug modellens eget gæt som
"nyeste dag" og gæt igen (og igen...). Se, hvad der sker:

```python
start_vindue = X_test_3d[0]                      # de første 7 testdage (standardiseret)
prognose = []

vindue_nu = start_vindue.clone()
with torch.no_grad():
    for dag in range(14):
        gaet = lstm_model(vindue_nu.unsqueeze(0)).squeeze()      # gæt næste dag
        prognose.append(gaet.item() * spredning + gns)
        # glid vinduet: smid ældste dag ud, sæt gættet ind bagerst
        vindue_nu = torch.cat([vindue_nu[1:], gaet.reshape(1, 1)])

plt.plot(range(14), y_test_np[:14], "o-", label="faktisk temperatur")
plt.plot(range(14), prognose, "s--", label="LSTM-prognose (gæt på gæt)")
plt.xlabel("dage ud i fremtiden")
plt.ylabel("°C")
plt.legend()
plt.title("Prognose 14 dage frem — fejlen vokser")
plt.show()
```

##### Opgave 9.3
Prøv hukommelses-størrelser på **4, 32 og 128** og sammenlign MAE (og fornemmelsen af
træningstid). Betaler en stor hukommelse sig på det her problem?

```python
for hukommelse in [4, 32, 128]:
    torch.manual_seed(42)
    m_h = VejrLSTM(hukommelse=hukommelse)
    optimizer = torch.optim.Adam(m_h.parameters(), lr=0.001)
    for epoke in range(60):
        for i in range(0, len(X_traen_3d), 256):
            optimizer.zero_grad()
            tab = tabsfunktion(m_h(X_traen_3d[i:i + 256]).squeeze(), y_traen[i:i + 256])
            tab.backward()
            optimizer.step()
    with torch.no_grad():
        f = m_h(X_test_3d).squeeze() * spredning + gns
    print(f"hukommelse {hukommelse}: {(f - y_test_celsius).abs().mean().item():.2f} grader")
```

<span style="color:red"><b>LØSNINGSFORSLAG:</b> Målt: hukommelse 4 → ~2,2 grader (for lille til overhovedet at lære opgaven på 60 epoker — underfitting), 32 og 128 → ~1,5 grader (ens, men 128 er mærkbart langsommere). Over en vis minimumsstørrelse er hukommelsen ligegyldig her, for der er ikke noget kompliceret at huske. Modellens kapacitet skal matche PROBLEMETS kompleksitet — endnu et eksempel på, at større ikke er bedre i sig selv.</span>

Giv LSTM'en en måneds hukommelse: byg vinduer med `vindue = 30` (genbrug opskriften fra

ekstra 23 dage?

```python
vindue30 = 30
Xl, yl = [], []
for i in range(len(temperatur) - vindue30):
    Xl.append(temperatur[i:i + vindue30])
    yl.append(temperatur[i + vindue30])
Xa, ya = np.array(Xl), np.array(yl)
n = int(0.8 * len(Xa))
g30, s30 = Xa[:n].mean(), Xa[:n].std()
Xt30 = torch.tensor((Xa[:n] - g30) / s30).unsqueeze(2)
Xv30 = torch.tensor((Xa[n:] - g30) / s30).unsqueeze(2)
yt30 = torch.tensor((ya[:n] - g30) / s30)

torch.manual_seed(42)
m30 = VejrLSTM()
optimizer = torch.optim.Adam(m30.parameters(), lr=0.001)
for epoke in range(60):
    for i in range(0, len(Xt30), 256):
        optimizer.zero_grad()
        tab = tabsfunktion(m30(Xt30[i:i + 256]).squeeze(), yt30[i:i + 256])
        tab.backward()
        optimizer.step()
with torch.no_grad():
    f = m30(Xv30).squeeze() * s30 + g30
print(f"LSTM med 30 dages vindue: {(f - torch.tensor(ya[n:])).abs().mean().item():.2f} grader")
```

<span style="color:red"><b>LØSNINGSFORSLAG:</b> Stadig ~1,5-1,6 grader — de ekstra 23 dage tilfører intet, for informationen om i morgen ligger (næsten) udelukkende i de seneste par dage. Bemærk at det IKKE er LSTM'ens skyld: den kan sagtens HÅNDTERE 30 skridt — der er bare intet at hente. Kend forskel på "modellen kan ikke" og "informationen findes ikke".</span>

##### Opgave 9.7 (ekstra)
Giv modellen flere sanser: byg vinduer med **to features pr. dag** — temperatur OG
luftfugtighed (kolonnen `fugt` i `dagligt_df`). LSTM'en skal så have
`input_size=2`. Hjælper fugtigheden på prognosen?

```python
fugt = dagligt_df["fugt"].values.astype("float32")   # luftfugtighed pr. dag (allerede resamplet)

Xl, yl = [], []
for i in range(len(temperatur) - 7):
    par = np.stack([temperatur[i:i + 7], fugt[i:i + 7]], axis=1)
    Xl.append(par)
    yl.append(temperatur[i + 7])
Xa, ya = np.array(Xl), np.array(yl)

n = int(0.8 * len(Xa))
g2 = Xa[:n].reshape(-1, 2).mean(axis=0)
s2 = Xa[:n].reshape(-1, 2).std(axis=0)
Xt2 = torch.tensor((Xa[:n] - g2) / s2)
Xv2 = torch.tensor((Xa[n:] - g2) / s2)
yt2 = torch.tensor((ya[:n] - g2[0]) / s2[0])

class VejrLSTM2(nn.Module):
    def __init__(self, hukommelse=32):
        super().__init__()
        self.lstm = nn.LSTM(input_size=2, hidden_size=hukommelse, batch_first=True)
        self.fc = nn.Linear(hukommelse, 1)

    def forward(self, x):
        hukommelser, _ = self.lstm(x)
        return self.fc(hukommelser[:, -1, :])

torch.manual_seed(42)
m2 = VejrLSTM2()
optimizer = torch.optim.Adam(m2.parameters(), lr=0.001)
for epoke in range(60):
    for i in range(0, len(Xt2), 256):
        optimizer.zero_grad()
        tab = tabsfunktion(m2(Xt2[i:i + 256]).squeeze(), yt2[i:i + 256])
        tab.backward()
        optimizer.step()
with torch.no_grad():
    f = m2(Xv2).squeeze() * s2[0] + g2[0]
print(f"LSTM med temperatur + fugtighed: {(f - torch.tensor(ya[n:])).abs().mean().item():.2f} grader")
```

<span style="color:red"><b>LØSNINGSFORSLAG:</b> Typisk en HÅRFIN forbedring (eller dødt løb) — fugtigheden bærer lidt information om vejrskift, men ikke meget på døgnskala. Det vigtige er mønstret: <code>input_size=2</code> og vinduer med shape (antal, 7, 2) — sådan skalerer man til 10 eller 100 features pr. tidsskridt. Rigtige prognosemodeller æder hundredvis af variable på præcis den her måde.</span>
