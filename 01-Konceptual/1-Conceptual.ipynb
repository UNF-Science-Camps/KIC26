# 1: Regression VS Klassifikation
I machine learning har vi mange forskellige modeller og arkitekturer som hver især er god til forskellige opgaver. De to typer opgaver man typisk lærer om ligger i hjertet af supervised learning – nemlig <strong>regression</strong> og <strong>klassifikation</strong>.
<br><br>
<strong>Regression</strong><br>
Som du nok kan huske, handler regression om at estimere en <strong>kontinuerlig</strong> værdi. Altså det her om at approksimere et tal som kan påtage sig mange forskellige værdier som kunne ligge på en slags linje eller bue.

<strong>Klassifikation</strong><br>
Klassifikation handler derimod om at differentiere (ikke som i d/dx) eller diskriminere (ikke som i den onde slags) mellem data punkter. Altså ville det svare til at sortere datapunkter ind i forskellige kasser baseret på deres features.

# 3: Parametre og Parametre Estimation
Machine learning er meget bygget på grundlaget af statistik. Faktisk kan man betragte ML modeller som mere fancy statistiske modeller idet de prøver at karakterisere problemer og udfald ud fra en funktion og en række karakteristiske parametre som typisk er ukendte.
<br><br>
Nedenunder er et par eksempler som understreger parametrene i nogle forskellige situationer så du selv kan få intuition i at genkende parametre i dine fremtidige modeller.
</div><br>

<br><hr style='max-width: 90%;'><br>

#### Eksempel 3.1
Du har opdaget en korrelation i antallet gange 6-7 siges om dagen til hvor mange gange arrangørerne rammer psykosen. Helt specifikt er disse to tal proportionelle, så når den ene vokser, vokser den anden med samme brøkdel af dens størrelse. Du er ML'er, så du har valgt at lave en model som kan forudse hvor mange gange arrangørerne rammer psykosen baseret på hvor mange gange der siges 6-7 om dagen.

$$\text{Psykosen} \propto \text{6-7} \;~\; \iff \text{Psykosen} = K \cdot \text{6-7} + \epsilon$$

Hvor epsilon er et lille fejl-led.
<br><br>
I denne simple model har vi kun én parameter, nemlig proportionalitetsparametren <strong>K ~ 0.067</strong>, da den er det eneste ukendte led som vi er interesserede i og ikke kender. Epsilon er ikke en parameter da vi antog at den var lille og tilfældig fra psykose til psykose, hvilket gør den ligegyldig.
</div><br>

## Hvorfor vi splitter vores data
Til at starte med, lad os glemme om valideringssættet og istedet fokusere kun på vores fulde dataset. Hvis du træner en model til at genkende mønstre i din data kan den typisk finde en løsning ret hurtigt og ret præcist med nok kompleksitet og lidt træning. 
<br><br>
Dette lyder godt, men det er virkelig ikke særlig imponerende fordi det ville svarer til at du giver en klog model din prøve og svar-arket og beder den om at finde svarene. Hurtigt ville din model finde ud af at den kan ignorere spørgsmålene helt og i stedet fokusere på kun at huske svar-arket, da det er det eneste du er interesseret i.
<br><br>
I stedet ville det være bedre hvis du baserede din models præstation på en ny prøve på samme emne, men med andre spørgsmål som den ikke har set før.
<br><br>
Når vi splitter vores data for første gang er dette hvad vi gør. Og på den måde stopper vi vores <strong>model</strong> fra at overfitte.
</div><br>

```python
# Eksempel 4.1: Sådan splitter du din data
import pandas as pd

# læs data ind
df = pd.read_csv('Pokemon.csv', index_col='#')
df
```

Vi har indlæst alle 800 Pokémon ind i en DataFrame. Nu vil vi så splitte dem op i et trænings- og testsæt.


```python
# manuel implementeringt
def split_data(df, test_size=0.2):
    n = len(df)

    # find split at 1 - test-size
    n_split = int((1-test_size) * n)

    df_train = df[:n_split]
    df_test = df[n_split:]

    return df_train, df_test

df_train, df_test = split_data(df, test_size=0.2)

# display er bare en fancy print
display(df_train)
display(df_test)
```

Heldigvis kan vi se at vores metode splitter vores 800 data punkter ind i et træningssæt af 0.80*800 = 640 punkter og 0.20*800 = 160 punkter, ergo har vi nu splittet vores data som man ville gøre i praksis.
<br><br>
Men nu hvor vi har set hvordan man implementerer det vil har set hvordan det implementeres vil vi aldrig implementere det igen og i stedet lade scikit-learn klare alt det hårde arbejde for os. Scikit-learns split-funktion er også bedre da du selv kan vælge om du vil blande din data eller stratificere den.


```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, test_size=0.2, random_state=1000)

display(df_train)
display(df_test)
```
